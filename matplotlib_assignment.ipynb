```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students, marks, color=['skyblue', 'lightgreen', 'orange', 'pink'])
plt.title('Student Marks')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()

```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1 if val == max(revenue) else 0 for val in revenue]
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode, startangle=90)
plt.title('Revenue Distribution by Region')
plt.show()


```

```python
data = np.random.randint(1, 101, 1000)
plt.hist(data, bins=10, edgecolor='black')
plt.title('Frequency Distribution of Random Integers (1–100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
