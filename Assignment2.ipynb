```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
trials = 10000
heads = 0
tails = 0

for _ in range(trials):
    toss = random.choice(['H', 'T'])
    if toss == 'H':
        heads += 1
    else:
        tails += 1

p_heads = heads / trials
p_tails = tails / trials

print("Heads:", heads, "Tails:", tails)
print(f"Experimental Probability of Heads = {p_heads:.3f}")
print(f"Experimental Probability of Tails = {p_tails:.3f}")
```

```python
trials = int(input("Enter number of trials: "))
count_sum7 = 0

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    if die1 + die2 == 7:
        count_sum7 += 1

p_sum7 = count_sum7 / trials
print(f"Experimental Probability of sum=7: {p_sum7:.3f}")
```

```python
def prob_at_least_one_six(trials=10000):
    success = 0
    for _ in range(trials):
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:
            success += 1
    return success / trials

p_one_six = prob_at_least_one_six()
print(f"Estimated Probability of at least one '6' in 10 rolls: {p_one_six:.3f}")
```

```python
colors = ['Red'] * 5 + ['Green'] * 7 + ['Blue'] * 8
draws = []
for _ in range(1000):
    draws.append(random.choice(colors))


count_prev_blue = 0
count_red_given_blue = 0

for i in range(1, len(draws)):
    if draws[i-1] == 'Blue':
        count_prev_blue += 1
        if draws[i] == 'Red':
            count_red_given_blue += 1

p_red_given_blue = count_red_given_blue / count_prev_blue if count_prev_blue > 0 else 0

print(f"Estimated P(Red | Blue previous) = {p_red_given_blue:.3f}")
```

```python
values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probabilities)

mean_empirical = np.mean(sample)
var_empirical = np.var(sample)
std_empirical = np.std(sample)

print(f"Empirical Mean = {mean_empirical:.3f}")
print(f"Empirical Variance = {var_empirical:.3f}")
print(f"Empirical Std Dev = {std_empirical:.3f}")
```

```python
mean = 5
samples = np.random.exponential(scale=mean, size=2000)


plt.figure(figsize=(8, 5))
count, bins, ignored = plt.hist(samples, bins=30, density=True, color='skyblue', alpha=0.7, label='Histogram')


x = np.linspace(0, 30, 200)
pdf = (1/mean) * np.exp(-x/mean)
plt.plot(x, pdf, 'r', linewidth=2, label='PDF Overlay')

plt.title("Exponential Distribution (mean = 5)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
uniform_data = np.random.uniform(0, 1, 10000)


n = 30
sample_means = [np.mean(np.random.choice(uniform_data, n)) for _ in range(1000)]


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(uniform_data, bins=30, color='orange', alpha=0.7)
plt.title("Original Uniform Distribution")

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='lightgreen', alpha=0.7)
plt.title("Distribution of Sample Means (n=30)")

plt.tight_layout()
plt.show()
```
